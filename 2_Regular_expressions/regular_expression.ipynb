```python
emails = "raju@ab c_:1g.com"
uemails=re.sub(r"@\w+","@gmail",emails) 
#\w+ → match one or more word characters ("letters, digits, underscore")
print(uemails)

# the pattern matches these parts:
# @abc
# @pqr
# @vvv
```

```python
emails = "raju@abc_1g.com and ksr@pqr.com and vara@vvv.in"
uemails=re.sub(r"@\w+","@gmail",emails) 
#\w+ → match one or more word characters ("letters, digits, underscore")
print(uemails)

# the pattern matches these parts:
# @abc
# @pqr
# @vvv
```

```python
text = "Raju works at Google in India"
caps = re.findall(r"\b[A-Z]\w+", text)
print(caps)

# How it works
# \b → word boundary
# [A-Z] → capital letter
# \w+ → rest of word
```

```python
phone = "9876543210"
valid = bool(re.fullmatch(r"\d{10}", phone))
print(valid)
```

```python
import re

text = "100kg 50kg 30m"
nums = re.findall(r"\d+(?=kg)", text)
print(nums)

# Working
# \d+ → number
# (?=kg) → must be followed by kg
# kg not included in result
```

```python
text = "cat car cap can"
words = re.findall(r"ca(?!r)\w", text)
print(words)

# Working
# ca → literal
# (?!r) → next char NOT r
# \w → next letter
```

```python
text = "bye bye bye hi hi test ok"
dup = re.findall(r"\b(\w+)\s+\1", text)
print(dup)

# Working
# (\w+) → capture word
# \1 → same word again
# Output
```

```python
text = "raju@gmail.com"
m = re.search(r"@(\w+)\.", text)
print(m.group(1))

# Working
# @ literal
# (\w+) → domain captured
# \. → stop at dot
```

```python
pwd = "Abc@1234"
valid = bool(re.fullmatch(
    r"(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[@#$%]).{8,}",
    pwd))

print(valid)

# Regex parts
# (?=.*[A-Z]) → at least 1 capital
# (?=.*[a-z]) → at least 1 small
# (?=.*\d) → at least 1 digit
# (?=.*[@#$%]) → special char
# .{8,} → length ≥ 8
```
