```python
import re

line="pet:catI love cats"
mat=re.match(r"pet:\w\w\w\w",line)
print(mat)
print(mat.group(0))
```

```python
line="pet:catI love cats"
mat=re.findall(r"pet:\w\w\w\w",line)
print(mat)
```

```python
line="pet:catI love     cats "
result=re.split(r"\s",line)
print(result)
```

```python
line="pet   d  d"
result=re.split(r"\S",line)
print(result)
```

```python
emails = "raju@abc.com and ksr@pqr.com and vara@vvv.in"
uemails=re.sub(r"@\w+","@gmail",emails)
print(uemails)
```

```python
re.sub(r"\.(com|in)",
       lambda m:".in" if m.group(1)=="com" else ".com",
       emails)
```
