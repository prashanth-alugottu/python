```python
# 1. sns.heatmap(data, annot=True)
# Used to display correlation or pivot tables with annotation (values inside cells).
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample correlation matrix
data = pd.DataFrame({
    'Math': [88, 92, 80, 89, 100],
    'Science': [95, 85, 76, 88, 92],
    'English': [78, 90, 85, 80, 86]
})
print(data)
```

```python
corr_matrix = data.corr()

# Heatmap with annotations
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
# 2. sns.clustermap(data)
# Visualizes hierarchical clustering of rows & columns.
sns.clustermap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python

# 3. sns.corrplot(corr_matrix) 
# sns.corrplot doesnt exist directly in Seaborn  its usually done with 
sns.heatmap(corr_matrix)
# So, this is basically a custom correlation heatmap.


sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
plt.title("Custom Correlation Plot")
plt.show()
```

```python
# 4. sns.pairplot(data)
# Shows pairwise relationships in a dataset.
print(data)
sns.pairplot(data)
plt.show()
```

```python
# 5. sns.heatmap(data, cmap='coolwarm')
# Change colormap for emphasis.
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
# 6. sns.heatmap(data, linewidths=0.5)
# Add grid spacing lines for readability.

sns.heatmap(corr_matrix, annot=True, linewidths=0.5)
plt.show()
```

```python
# 7. sns.heatmap(data, vmin=0, vmax=1)
# Normalize values to fit within a range.
sns.heatmap(corr_matrix, annot=True, vmin=0, vmax=1, cmap="Blues")
plt.show()
```
