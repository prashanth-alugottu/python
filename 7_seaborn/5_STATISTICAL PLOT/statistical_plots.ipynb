```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = pd.DataFrame({
    "Maths": [45, 67, 78, 56, 89, 92, 58, 76, 88, 95],
    "Science": [40, 65, 80, 55, 86, 90, 60, 74, 85, 97],
    "Gender": ["M","F","M","F","M","M","F","M","F","M"]
})
print(data)
```

```python
# 1. Regression Plot (sns.regplot)
# Scatterplot + regression line
sns.regplot(x="Maths", y="Science", data=data)
plt.show()
```

```python
# 2. Linear Model Plot (sns.lmplot)
# Scatterplot + regression line with grouping (hue)
sns.lmplot(x="Maths", y="Science", hue="Gender", data=data)
plt.show()
```

```python
# 3. KDE Plot (sns.kdeplot)
# Density visualization (bivariate)
sns.kdeplot(x="Maths", y="Science", data=data, cmap="Blues", fill=True)
plt.show()
```

```python
# 4. Rug Plot (sns.rugplot)
# Small ticks along the axis
sns.rugplot(x=data["Maths"], height=0.05)
plt.show()
```

```python
# 5. Joint Plot (KDE)
# Joint distribution + KDE
sns.jointplot(x="Maths", y="Science", data=data, kind="kde")
plt.show()
```

```python
# 6. Joint Plot (Regression)
# Joint distribution + regression line
sns.jointplot(x="Maths", y="Science", data=data, kind="reg")
plt.show()
```

```python
# 7. Boxen Plot (sns.boxenplot)
# Enhanced boxplot for large datasets
sns.boxenplot(x="Gender", y="Maths", data=data)
plt.show()
```

```python
# 8. ECDF Plot (sns.ecdfplot)
# Cumulative distribution function
sns.ecdfplot(data["Maths"], complementary=True)
plt.show()
```
