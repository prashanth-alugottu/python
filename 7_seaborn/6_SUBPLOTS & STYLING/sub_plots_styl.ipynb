```python
# 1. sns.set_style('whitegrid') → Apply grid-based styling for readability
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample data
tips = sns.load_dataset("tips")
print(tips)
sns.set_style("whitegrid")  # Apply grid background
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Boxplot with Whitegrid Style")
plt.show()
# Makes background cleaner with grid lines → easier to compare values.
```

```python
# 2. sns.set_palette('pastel') → Change color palettes for consistency
sns.set_style("whitegrid")
sns.set_palette("pastel")  # Soft pastel colors
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Barplot with Pastel Palette")
plt.show()
# Helps maintain visual consistency across multiple plots.
```

```python
# 3. sns.despine() → Remove chart spines (top/right borders)
sns.set_style("whitegrid")
sns.barplot(x="sex", y="total_bill", data=tips)
sns.despine()  # Removes top & right spines
plt.title("Barplot without Extra Spines")
plt.show()
# Gives charts a minimal and modern look.
```

```python
# 4. sns.set_context('talk') → Adjust figure scaling for presentations
sns.set_context("talk")  # Bigger fonts, thicker lines for presentation
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day")
plt.title("Scatterplot in 'Talk' Context")
plt.show()
# Contexts: "notebook", "talk", "paper", "poster" → Auto-adjust sizes for different use cases.
```

```python
# 5. sns.catplot(x, y, data, col='category') → Create facet grids for subcategories
sns.catplot(x="sex", y="total_bill", col="day",kind="box", data=tips)
# Creates multiple subplots automatically, one for each day.Very useful for comparing across categories.

```

```python
# 6. sns.relplot(x, y, col='category', data) → Subplot relational data by categories
sns.relplot(x="total_bill", y="tip", col="time",
            hue="day", data=tips)
# Breaks scatterplots into separate subplots for Lunch vs Dinner.
```

```python
# 7. sns.set_theme(style='dark') → Apply Seaborn themes
sns.set_theme(style="dark")  # Dark theme applied
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.title("Histogram with Dark Theme")
plt.show()
# Themes: "dark", "white", "ticks", "darkgrid", "whitegrid".
```

```python
# 8. sns.color_palette('deep', as_cmap=True) → Set & reuse color palettes
palette = sns.color_palette("deep", as_cmap=True)
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, palette="deep")
plt.title("Scatterplot with Deep Palette")
plt.show()
# Ensures consistent colors across all plots.
```
