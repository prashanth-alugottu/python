```python
import seaborn as sns
import matplotlib.pyplot as plt
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

```python
tips=sns.load_dataset("tips")
print(tips)
# Histogram of 'total_bill'
sns.histplot(tips['total_bill'],bins=30)
plt.show()
```

```python
print("2. KDE Plot (Kernel Density Estimate)")
sns.kdeplot(tips['total_bill'], fill=True)
plt.show()
# Smooth curve showing distribution (probability density)
```

```python
print("3. Rug Plot")
sns.rugplot(tips['total_bill'])
plt.show()
# Adds small tick marks on the axis for each data point.
```

```python
print("4. Displot (Deprecated but useful)")
sns.displot(tips['total_bill'], bins=30, kde=True)
plt.show()
#  Combines Histogram + KDE in one plot.
```

```python
print("5. ECDF Plot (Empirical Cumulative Distribution Function)")
sns.ecdfplot(tips['total_bill'])
plt.show()
# Shows cumulative percentage of data points.
```

```python
print("6. Violin Plot")
sns.violinplot(x=tips['day'], y=tips['total_bill'])
plt.show()
# Shows distribution + KDE + boxplot inside.
```

```python
print("7. Box Plot")
sns.boxplot(x=tips['day'], y=tips['total_bill'])
plt.show()
# Shows median, quartiles, and outliers.
```

```python
print("8. Joint Plot")
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
# Bivariate distribution with hexagonal bins.
```
