```python
    # 1. plt.savefig('plot.png')
# Save plots as PNG images.
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.savefig("scatter.png")   # saves in PNG format
plt.show()
```

```python
# 2. plt.savefig('plot.pdf', dpi=300)
# Export plots in high-resolution PDF, good for printing.
sns.histplot(tips["total_bill"], bins=20)
plt.savefig("histogram.pdf", dpi=300)   # high-resolution
plt.show()
```

```python
# 3. plt.savefig('plot.svg', transparent=True)
# Save as SVG (vector format), scalable without losing quality.
sns.boxplot(x="day", y="total_bill", data=tips)
plt.savefig("boxplot.svg", transparent=True)  # publication-ready
plt.show()
```

```python
# 4. plt.savefig('plot.jpeg', quality=95)
# Save plots as JPEG with custom quality.
sns.barplot(x="day", y="tip", data=tips)
plt.savefig("barplot.jpeg", quality=95)   # high-quality JPEG
plt.show()
```

```python
# 5. plt.savefig('plot.png', bbox_inches='tight')
# Save figure with minimal whitespace.
sns.violinplot(x="day", y="tip", data=tips)
plt.savefig("violin.png", bbox_inches="tight")  # trims extra space
plt.show()
```

```python
# 6. sns.despine()
# Remove extra borders (spines) before saving.
sns.boxplot(x="day", y="tip", data=tips)
sns.despine()  
plt.savefig("clean_boxplot.png")
plt.show()
```

```python
# 7. sns.set_context('notebook')
# Adjust plot size for notebooks before saving.
sns.set_context("notebook")
sns.histplot(tips["tip"], bins=10)
plt.savefig("notebook_hist.png")
plt.show()
```

```python
# 8. plt.show()
# Always display and validate the plot before saving.
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips)
plt.show()   # check before saving
```
