```python
try:
    with open("MyFiles.txt",mode='r',encoding="utf-8") as myfile:
        # mode = r --> read only cannot write
        # mode = r+ --> read and write
        # print(myfile.mode)
        # print(myfile.readable())
        # print("is writtable : ",myfile.writable())
        # print(myfile.write("Hi Prashanth Algottu, we’ve curated job opportunities where you’d be a top applicant, based on your profile and preferences."))
        # myfile.writelines(""" 
        #                   Hi Prashanth 
        #                   Hello Python
        #                   """)
        
        # Reading 
        print("After reading : ",myfile.read(5))
        print("All line : ",myfile.readlines())
        print("Length of all liness : ",len(myfile.readlines()))
        
        # cursore position
        mycursorPos=myfile.tell()
        print("Before cursor ",mycursorPos)
        newCurPos=myfile.seek(0) # read from the start using myfile obj not with mycursorPos obj
        print("After reset the cursor position",newCurPos)
        print(myfile.read(5))
except IOError:
    print("File not found")
finally:
    print("Final Block")
```

```python
# creating csv ==> comma separated values
import csv

try:
    with open("myfile.csv","w",newline="") as mycsv:
        a=csv.writer(mycsv,delimiter=",")
        data=[['STOCK','SALES','PRICE'],
              ['100','90','90$'],
              ['100','90','90$'],
              ['100','90','90$'],
              ['100','90','90$'],]
        a.writerows(data)
        print("Suuceess")
except IOError:
    print("File not finding")
finally:
    print("Finally block")        
        
```

```python
# TASK -1 Appending text file
try:
    with open("combined.txt",mode='w') as out:
        myfiles=["Myfiles1.txt","MyFiles.txt"]
        for file in  myfiles:
            with open(file,mode='r') as f:
                out.write(f.read())
except IOError:
    print("File not found")
finally:
    print("Finallly.  ")
```
