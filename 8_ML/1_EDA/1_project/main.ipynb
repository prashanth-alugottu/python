```python
df_country=pd.read_excel("Zomatodataset/Country-Code.xlsx")
df_country
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When Rating is between 4.5 to 4.9 --> Excellent
2. When Ratings are between 4.0 to 4.5 --> Very Good
3. When Ratings are between 3.5 ro 3.9 --> Good
4. When Ratings are between 3.0 ro 3.5 --> Average
5. When Ratings are between 2.5 ro 2.9 --> Average
6. When Ratings are between 2.0 ro 2.4 --> Poor


```python
ratings.head()
```

```python
import matplotlib as mpl
mpl.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green'])
```
