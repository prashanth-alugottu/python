# **Exploratory Data Analysis (EDA) â€” Iteration 2**  
### *Univariate, Bivariate, and Multivariate Analysis with Professional Insights*

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset (Iris)
df = sns.load_dataset("iris")

# Display first 5 rows
df.head()
```

## **Step 1 â€” Univariate Analysis**
ðŸ“Œ **Purpose:**  
Analyze one variable at a time to understand its distribution, spread, and possible outliers.  

Weâ€™ll cover:  
- **Numerical variables:** Histogram & Boxplot  
- **Categorical variables:** Countplot & Pie Chart  


#### **Univariate: Numerical :** Histogram & Boxplot

```python
# Numerical columns
num_cols = df.select_dtypes(include='number').columns

# Histogram for numerical columns
for col in num_cols:
    plt.figure(figsize=(6,4)) # We can add subplot
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.show()
```

#### kde=True â†’ Adds a smooth density curve.


```python
# Boxplot for numerical columns
for col in num_cols:
    plt.figure(figsize=(6,4))  # We can add subplot
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

#### Boxplots are quick to detect outliers visually.

#### **Univariate: Categorical :**  Countplot & Pie Chart 

```python
# Categorical columns
cat_cols = df.select_dtypes(include='object').columns

# Countplot for categorical columns
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=df[col], palette='Set2')
    plt.title(f'Count of {col}')
    plt.show()
```

```python
# Pie chart for categorical columns
for col in cat_cols:
    plt.figure(figsize=(6,4))
    df[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
    plt.title(f'Pie Chart of {col}')
    plt.ylabel("")
    plt.show()
```

#### autopct='%1.1f%%' â†’ shows percentages in pie chart.

#### startangle=90 â†’ makes pie charts start from the top for symmetry.

## **Step 2 â€” Bivariate Analysis**
ðŸ“Œ **Purpose:**  
Examine the relationship between two variables.

Weâ€™ll cover:  
- **Numerical & Categorical:** Boxplot  
- **Categorical & Categorical:** Countplot with hue  
- **Numerical & Numerical:** Scatterplot, LM plot  


#### **Numerical & Categorical**

```python
# Example: Petal length by species
sns.boxplot(x='species', y='petal_length', data=df, palette='Set3')
plt.title('Petal Length by Species')
plt.show()
```

#### **Categorical & Categorical**

```python
# Create petal width category
df['petal_width_cat'] = pd.cut(df['petal_width'], bins=[0,0.5,1.5,2.5], labels=['Small','Medium','Large'])

# Countplot with hue
sns.countplot(x='species', hue='petal_width_cat', data=df, palette='coolwarm')
plt.title('Species vs Petal Width Category')
plt.show()
```

#### **Numerical & Numerical**

```python
# Scatter plot
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df, palette='viridis')
plt.title('Sepal Length vs Petal Length')
plt.show()

# Regression plot
sns.lmplot(x='sepal_length', y='petal_length', hue='species', data=df, height=5, aspect=1.2)
```

#### lmplot adds regression lines to detect trends.

#### Scatter plots show clusters & relationships.

## **Step 3 â€” Multivariate Analysis**
ðŸ“Œ **Purpose:**  
Study relationships between 3 or more variables at the same time.  
Weâ€™ll use **Pairplot** and **Correlation Heatmap**.  


```python
# Pairplot
sns.pairplot(df, hue='species', palette='husl')
plt.show()
```

### * Pairplots are great for detecting clusters and correlations.

```python
# Correlation heatmap (only numeric columns)
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Heatmaps are used for feature selection in ML.
