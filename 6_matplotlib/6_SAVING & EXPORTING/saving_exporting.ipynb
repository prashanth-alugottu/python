```python
import matplotlib.pyplot as plt
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 11]
plt.plot(x, y, marker='o', color='blue')
plt.title("Sales Trend")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.savefig('plot.png')
```

```python
print("6.Save with Tight Layout (no extra space)")
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 11]
plt.plot(x, y, marker='o', color='blue')
plt.title("Sales Trend")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.savefig('plot2.png', bbox_inches='tight')
```
