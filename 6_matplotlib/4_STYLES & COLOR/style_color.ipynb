```python

import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 5, 7, 1, 6]
data = [7, 8, 5, 6, 2, 3, 9, 4, 5, 7]
sizes = [30, 25, 20, 15, 10]
```

```python
print("1. plt.style.use('ggplot') → Apply style theme")
plt.style.use('ggplot')
plt.plot(x, y)
plt.title("ggplot Style")
plt.show()
```

```python
print("2. plt.plot(x, y, 'ro') → Style with markers")
# (Red circles 'ro' → r=red, o=circle marker)
plt.plot(x, y, 'ro')
plt.title("Red Circle Markers")
plt.show()
```

```python
print("3. plt.plot(x, y, linestyle='--', color='g') → Green dashed line")
plt.plot(x, y, linestyle='--', color='g')
plt.title("Green Dashed Line")
plt.show()
```

```python
print("4. plt.scatter(x, y, c='blue', marker='x') → Scatter with style")
plt.scatter(x, y, c='blue', marker='x')
plt.title("Scatter with Blue X")
plt.show()
```

```python
print("5. plt.bar(x, y, color='skyblue') → Bar chart colors")
plt.bar(x, y, color='skyblue')
plt.title("Skyblue Bar Chart")
plt.show()
```

```python
print("6. plt.hist(data, bins=9, color='purple') → Histogram colors")
plt.hist(data, bins=9, color='purple')
plt.title("Purple Histogram")
plt.show()
```

```python
print("7. plt.plot(x, y, alpha=0.5) → Transparency (50% faded)")
plt.plot(x, y, color='red', alpha=0.5, linewidth=3)
plt.title("Transparent Line (alpha=0.5)")
plt.show()
```

```python

print("8. plt.pie(sizes, colors=['red','green','blue','orange','purple']) → Pie chart with colors")
plt.pie(sizes, colors=['red','green','blue','orange','purple'], autopct='%1.1f%%')
plt.title("Pie Chart with Colors")
plt.show()
```
