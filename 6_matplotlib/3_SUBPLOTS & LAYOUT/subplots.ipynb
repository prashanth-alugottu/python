```python
import matplotlib.pyplot as plt

print('1.Matplotlib allows us to create multiple plots in the same figure using plt.subplot() or plt.subplots().')


# 2 rows, 2 columns, 1st plot
plt.subplot(2,2,2) # size
plt.plot([1,2,3],[4,5,6],label="Square")
plt.title("Plot 1")

# 2nd plot
plt.subplot(2, 2, 2)
plt.plot([1,2,3], [1,2,3], label="Line")
plt.title("Plot 2")

plt.tight_layout()
plt.legend()
plt.show()
```

```python
print("2. fig, ax = plt.subplots() (Better way)")
fig,ax=plt.subplots()
print(fig)
ax.plot([1,2,3], [1,4,9])
ax.set_title("Simple Subplot")
plt.show()
```

```python
print("3. Multiple subplots in grid (2Ã—2)")

# Easier layout with plt.subplots(rows, cols).

fig, ax = plt.subplots(3, 3)
ax[0,0].plot([1,2,3],[1,4,9])
ax[0,0].set_title("Square")

ax[0,1].plot([1,2,3],[1,2,3])
ax[0,1].set_title("Line")

ax[0,2].plot([1,2,3],[1,2,3])
ax[0,2].set_title("Line")

ax[1,0].bar([1,2,3],[3,2,5])
ax[1,0].set_title("Bar")

ax[1,1].hist([1,1,2,3,3,3,4,5])
ax[1,1].set_title("Histogram")

ax[1,2].hist([1,1,2,3,3,3,4,5])
ax[1,2].set_title("Histogram")



plt.tight_layout()
plt.show()
```

```python
print("4. Sharing axes (sharex, sharey)")

# Useful for comparing plots.

fig,ax=plt.subplots(2,sharex=False)
ax[0].plot([1,2,3], [1,4,9])
ax[0].set_title("Square")

ax[1].plot([100,200,500], [1,2,3])
ax[1].set_title("Line")

plt.tight_layout()
plt.show()
```

```python
print("5. Adjust spacing with plt.tight_layout()")
# Avoids overlapping of titles and labels.
fig, ax = plt.subplots(2, 2)
for i in range(2):
    for j in range(2):
        ax[i,j].plot([1,2,3],[1,2,3])
        ax[i,j].set_title(f"Plot {i},{j}")
plt.tight_layout()
plt.show()
```

```python
print("6. figsize=(width, height)")
# Control size of figure.
fig, ax = plt.subplots(1, 2, figsize=(10,4))
ax[0].plot([1,2,3],[1,4,9])
ax[0].set_title("Wide")
ax[1].plot([1,2,3],[1,2,3])
ax[1].set_title("Plot 2")
plt.show()
```

```python

print("7. Adding super title (fig.suptitle)")
# Title for the entire figure.
fig, ax = plt.subplots(1, 2)
ax[0].plot([1,2,3],[1,4,9])
ax[1].plot([1,2,3],[1,2,3])
fig.suptitle("Comparison of Graphs", fontsize=14)
plt.show()

```

```python
print("8. Complex layout using GridSpec")
# More flexible than simple rows Ã— cols.
import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(6,6))
gs = gridspec.GridSpec(2,2)
ax1 = fig.add_subplot(gs[0,:])  # Full top row
ax2 = fig.add_subplot(gs[1,0])  # Bottom left
ax3 = fig.add_subplot(gs[1,1])  # Bottom right
ax1.plot([1,2,3],[1,4,9])
ax2.bar([1,2,3],[3,2,5])
ax3.hist([1,1,2,2,3,3,4,5])
plt.show()
```
