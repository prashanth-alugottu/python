```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

print(x)
print(y)
print("========================= 1. Line Plot =========================")
plt.plot(x,y,marker='o')
plt.title("Line plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

print("X-values ",x)
print("Y-values ",y)
print("========================= 2. Scatter Plot =========================")
plt.scatter(x,y,color='red')
plt.title("Scatter plot")
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 9]

print("categories ",categories)
print("values ",values)
print("========================= 3. Bar Chart =========================")
plt.bar(categories,values,color='skyblue')
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 2, 9]

print("categories ",categories)
print("values ",values)
print("========================= 4. Horizontal Bar Chart =========================")
plt.barh(x,y,color='skyblue')
plt.title("Horizontal bar chart")
plt.show()
```

```python
import numpy as np
data = np.random.randn(100)  # 1000 random numbers
print("Data :  ",data)
print("\n========================= 5. Histogram =========================")

plt.hist(data,bins=20,color='purple',edgecolor='black')
plt.title("Histogram")
plt.show()
```

```python

sizes = [30, 20, 25, 25]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']

print("\n========================= 6. Pie Chart =========================")

plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=140)
plt.title("Pie chart")
plt.show()
```

```python
data = [7, 8, 5, 6, 7, 8, 5, 10, 12, 6, 8, 7]

print("\n========================= 7. Box Plot =========================")
plt.boxplot(data)
plt.title("Box plot")
plt.show()
```

```python
import numpy as np
data = [np.random.normal(0, std, 5) for std in range(1, 4)]
print(data)
print("\n========================= 8. Violin Plot =========================")
plt.violinplot(data, showmeans=True)
plt.title("Violin Plot")
plt.show()
```
