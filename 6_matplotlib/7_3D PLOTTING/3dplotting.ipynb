```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # important

x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # create 3D axis

ax.plot3D(x, y, z, color='blue')

plt.show()
```

```python
print("5. Create a 3D Scatter Plot")
# Creates a **3D scatter plot** to show individual data points.
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x, y, z, color='red')
plt.show()
```
