```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

print("1. plt.annotate()")
# Used to add annotations with optional arrows pointing to specific points.

plt.plot(x, y, marker='o')
plt.annotate("Important Point", xy=(2, 4), xytext=(3, 5),
             arrowprops=dict(arrowstyle="->"))
plt.show()
```

```python
print("2. plt.text()")
# Places simple text at given coordinates.
plt.plot([1, 2, 3], [2, 4, 6])
plt.text(2, 4, "Mid Point")
plt.show()
```

```python
print("6. plt.xticks()")
# Customizes the tick values and labels on the x-axis.
plt.plot([0, 1, 2], [10, 20, 30])
plt.xticks([0, 1, 2], ['A', 'B', 'C'])
plt.show()
```

```python
print("7. plt.yticks()")
# Customizes the tick values and labels on the y-axis.
plt.plot([1, 2, 3], [10, 20, 30])
plt.yticks([10, 30], ['Low', 'High'])
plt.show()
```

```python
print("8. Simple Annotation using plt.text()")
# Adding basic annotation text inside the plot.
plt.plot([1, 2, 3], [2, 3, 5])
plt.text(2, 3, "Annotation")
plt.show()
```
