```python
import matplotlib.pyplot as plt


print("\n============== 4. plt.legend() → Show Legend ==============")
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, label="Line A")  
plt.plot(x, [i**2 for i in x], label="Line B")  
plt.legend()   # Display legend
plt.show()
```

```python

print("\n============== 5. plt.grid(True) → Show Gridlinesd ==============")
plt.plot(x, y)
plt.grid(True)   # Adds grid for better readability
plt.show()

```

```python
print("\n============== 6. plt.xlim([xmin, xmax]) → Set X-axis Limit ==============")
plt.plot(x, y)
plt.xlim([0, 6])   # Restrict x-axis from 0 to 6
plt.ylim([0, 12])  # Restrict y-axis from 0 to 12
plt.show()
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
sales = [100, 120, 90, 140, 160]
plt.bar(days, sales)
plt.xticks(rotation=45)   # Rotate x labels
plt.title("Sales Over Days")
plt.show()
```
