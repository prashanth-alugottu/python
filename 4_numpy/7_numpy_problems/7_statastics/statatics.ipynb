```python
import numpy as np

arr = np.array([3,4])
print(arr)
```

```python
mean=np.mean(arr)
print(mean)
```

```python
arr1=np.array([[1,2],[3,4],[5,6]])
print(arr1)
mean1=np.mean(arr1)
print(mean1)
```

```python
median=np.median(arr)
print(median)
print()
median1=np.median(arr1)
print(median1)

arr3=np.array([1,2,3])
median3=np.median(arr3)
print(median3)
```

```python
print(arr1)
std1=np.std(arr1)
print(std1)
print()

print(arr3)
std3=np.std(arr3)
print(std3)
print()


```

```python
print(arr1)
sum1=np.sum(arr1)
print(sum1)
print()

print(arr3)
sum3=np.sum(arr3)
print(sum3)
print()
```

```python
arr=np.array([1,33,63,38,77,9,0,33])
print(arr)
print()
print(np.max(arr))
```

```python
print(np.min(arr))
```

```python
print(np.percentile(arr,50))
print(np.percentile(arr,0))
print(np.percentile(arr,60))
print(np.percentile(arr,100))
```

```python
print(np.var(arr)) # Variance
```

```python
#  Correlation coefficient between two arrays
arr1=np.array([1,2,3,4])
arr2=np.array([5,6,7,8])
print(np.corrcoef(arr1,arr2))
```
