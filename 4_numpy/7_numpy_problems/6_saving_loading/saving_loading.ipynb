```python
import numpy as np
arr=np.array([[1,2],[3,4]])
print(arr)
```

```python
np.save('file.npy',arr)
```

```python
loadedfile=np.load('file.npy')
print(loadedfile)
```

```python
np.savetxt('arr.txt',arr)
```

```python
load=np.loadtxt('arr.txt')
print(load)
```

```python
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[10,20],[30,40]])
arr3=np.array([[11,21],[31,41]])
arr4=np.array([[12,22],[32,42]])

np.savez('fullfile',arr1,arr2,arr3,arr4)
```

```python
arrs=np.load('fullfile.npz')
print(arrs)
for keys in arrs:
    print(arrs[keys])
    print()
```

```python
np.savez_compressed('compression',arr1,arr2)
```

```python
arrs=np.load('compression.npz')
print(arrs)

for key in arrs:
    print(arrs[key])
    print()
```
