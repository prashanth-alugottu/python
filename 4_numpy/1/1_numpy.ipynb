```python
# dtype='i' it convert the array into "integer"
nparr = np.array([1,2,3,4,5],dtype='i')
print(nparr)

# dtype='S' it convert the array into 'string'
nparr = np.array([1,2,3,4,5],dtype='S')
print("Here 'b' represents bytes string : ",nparr)
# NOTE: For i, u, f, S and U we can define size as well. Here 'b' represents bytes string.

# dtype='U' it convert the array into "unicode string"
nparr = np.array([1,2,3,4,5],dtype='U')
print(nparr)

# dtype='U' it convert the array into "complex float"
nparr = np.array([1,2,3,4,5],dtype=complex)
print(nparr)
```

```python
# Converting Data Type on Existing Arrays
npar=np.array([1,2.3,4,6])
print("array float : ",npar)
print(npar.dtype)

# conveting float to int
intar=npar.astype('i') # 
print("converted to int : ",intar)
print(intar.dtype)
print("Older array type will not change : ",npar.dtype)

# conveting float to str
strar=npar.astype('S')
print("Converted to string and its a byte string that why 'b' is there : ",strar)
print(strar.dtype)

# conveting float to bool
exarr=np.array([0,1,-1,1.2])
barr=exarr.astype(bool)
print("Bool array : ",barr)


```
