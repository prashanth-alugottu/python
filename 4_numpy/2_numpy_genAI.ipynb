```python
# Python NumPy Array Creation from Numerical Ranges
# The Numpy library provides some functions to create an array from the given specified range. 
# 1    numpy.arange
# 2    numpy.linspace
# 3    numpy.logspace
```

```python
pyarr=np.arange(10)
print(pyarr)
```

```python
pyarr=np.arange(0,7,2,'complex')
print(pyarr)
```

```python
# 2. Using numpy.linspace
# It is similar to the arange() function. But there is a major difference and it is in this function, instead of step size, the number of evenly spaced values between the interval is specified using the num argument.

# Syntax:
# numpy.linspace(start, stop, num,endpoint, retstep,dtype)   

# start    => start of interval range. 
# stop     =>  end of interval range
# num      => [int, optional] No. of samples to generate, Default 50
# endpoint => If True, stop index is included, If False Not Included. Default, True
# ReturnStep   => It returns the difference between values. True-Visible, False-NotVisible(Default)
# dtype    => type of output array
```

```python
pyarr=np.linspace(0,10,4)
print(pyarr)
```

```python
PyArr = np.linspace(0, 10, 5)  
print(PyArr)
PyArr = np.linspace(0, 10, 3)  
print(PyArr)
PyArr = np.linspace(10, 20, 3,endpoint=False)  #EndIndexNotIncluded
print(PyArr)
PyArr = np.linspace(0, 10, 5,endpoint=False,retstep=True)  
print(PyArr)
PyArr = np.linspace(0, 10, 5,endpoint=False,retstep=True,dtype=int)  
print(PyArr)
```

```python

# 3. Using numpy.logspace
# It is used to create an array by using the numbers that are evenly separated on a log scale.

# Syntax:
# numpy.logspace(start, stop, num, endpoint, base, dtype) 
# start    => start of interval range. 
# stop     =>  end of interval range
# num      => [int, optional] No. of samples to generate, Default 50
# endpoint => If True, stop index is included, If False Not Included.Default, True
# base     => [float, optional] Base of log scale. Default-10.0
# dtype    => type of output array
```

```python
PyData = np.logspace(0, 10, num = 6, endpoint = True)  
print(PyData)
```

```python
# eye() function
# It is used to create a 2-D array with ones on the diagonal and zeros elsewhere.
# take square matrix, initialize 1's on all diagonal positions, 0's on all Non-diagonal positions.

# Syntax:
# numpy.eye(N, M=None, k=0, dtype='float')

# N	Number of rows in the output.			
# M	Number of columns in the output. If None, defaults to N.	
# k	Index of the diagonal: 0
# dtype	Data-type of the returned array.
```

```python
import numpy as np
print(np.eye(2))
print(np.eye(3,3))
print(np.eye(3,3,dtype=int))
```

```python
import numpy as np
print(np.eye(2,k=-1))
print(np.eye(2,2,k=-1))
print(np.eye(3,3,k=-2))
# print(np.eye(3,3,k=2))
```
