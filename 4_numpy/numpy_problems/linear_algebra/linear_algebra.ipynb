```python
import numpy as np

arr1=np.array([[1,2],[3,4]])

arr2=np.array([[5,6],[7,8]])

print(arr1)
print()
print(arr2)
print()

arr3=np.dot(arr1,arr2)
arr4=np.matmul(arr1,arr2)
print(arr3)
print(arr4)
print(arr1@arr2)
```

```python
arr2=np.array([[5,6],[7,8]])
print(arr2)
print()
print(arr2.T)
```

```python
arr=np.array([[1,2],[3,4]])
print(arr)
arr1=np.linalg.inv(arr)
print()
print(arr1)
```

```python
arr1=np.array([[1,2],[3,4]])
arr2=np.linalg.det(arr1)
print()
print(arr2)
```

```python
arr1=np.array([[1,2],[2,4]])
print(arr1)
print()
arr2=np.linalg.det(arr1)
print(arr2)
```

```python
print(arr)
eigvalues,eigvector=np.linalg.eig(arr)
print()
print(eigvalues)
print()
print(eigvector)
```

```python
print(arr)
print()
U,S,Vh=np.linalg.svd(arr)
print(U)
print()
print(S)
print()
print(Vh)
print()

```

```python
print(arr)
print()
norms=np.linalg.norm(arr)
print(norms)
```
