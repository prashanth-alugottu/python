```python
import numpy as np
arr = np.array([1, 2, 2, 3, 4, 4, 5,"chinnu","chinnu","Chinnu"])
arr1=np.unique(arr)
print(arr1)
```

```python
arr=[100,123,2,4,2,45,7,85,44]
arr2=np.sort(arr)
print(arr2)

arr=['chinnu','Chinnu','chanti',]
arr2=np.sort(arr)
print(arr2)

arr=['7','44', '45', '7', '85']
arr4=np.sort(arr)
print(arr4)
```

```python
arr=[100,123,2,4,2,45,7,85,44]
argsort=np.argsort(arr)
print(argsort) # Index sorting.
```

```python
arr=[100,123,2,4,2,45,7,85,44]
count=np.count_nonzero(arr)
print(count)

arr=[100,0,2,4,2,45,0,85,44]
count=np.count_nonzero(arr)
print(count)
```

```python
arr=[100,np.nan,2,np.nan,7,0]
nan=np.nan_to_num(arr)
print(nan)
nan=np.nan_to_num(arr,nan=1)
print(nan)
```

```python
print(arr)
nanarr=np.isnan(arr)
print(nanarr)
```

```python
arr=[100,np.nan,2,np.nan,7,0]
new_arr=np.delete(arr,1)
print(new_arr)
new_arr=np.delete(arr,0)
print(new_arr)
```
