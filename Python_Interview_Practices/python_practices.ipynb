```python
# 8️⃣ Running Difference
# Problem:Return a list where each element is the difference between current and previous element.
# Example:[5, 10, 7] → [0, 5, -3]
            
def current_pre_diff(nums):
    prev = nums[0]
    result=[]
    for num in nums:
        value=num-prev
        result.append(value)
        prev=num
    
    print(result)

current_pre_diff([5, 10, 7])
              
```

```python
def sliding_window(text, window_size=50, step=25):

    chunks = []
    for i in range(0, len(text), step):
        chunk = text[i:i+window_size]
        chunks.append(chunk)

    return chunks


text = "Artificial Intelligence is transforming industries and improving automation."

chunks = sliding_window(text)

for c in chunks:
    print(c)
```

```python
def is_palindrome(s):
    left, right = 0, len(s) - 1

    while left < right:
        print("s left : ", s[left])
        print("s right : ", s[right])
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True
print(is_palindrome("agpge"))

```

```python
def insertion_sort(arr):
    n = len(arr)

    for i in range(1, n): # n=8
        key = arr[i] # 4   # element to insert
        j = i - 1 # 1-1 = 0
            # j=0
        # shift elements greater than key
        while j >= 0 and arr[j] > key: # 6>4
            arr[j + 1] = arr[j] #4 = 6
            j -= 1 # 0-1 = -1

        # place key at correct position
        arr[j + 1] = key

    return arr


# Example
arr = [6, 4, 3, 6, 67, 3, 56, 5]
print(insertion_sort(arr))
```

```python
import copy
a = [1, 2, [3, 4]]
# Shallow Copy
sh_copy=copy.copy(a)
sh_copy[0]=10
print("original : ",a)
print("shallow : ",sh_copy)
print("=========================")
sh_copy[2][0]=30
print("original : ",a)
print("shallow : ",sh_copy)

```

```python
import copy
a = [1, 2, [3, 4]]
# Deep Copy
dp_copy=copy.deepcopy(a)
dp_copy[0]=10
print("original : ",a)
print("deep : ",dp_copy)
print("=========================")
dp_copy[2][0]=30
print("original : ",a)
print("deep : ",dp_copy)

# “Shallow copy shares inner objects, 
# deep copy creates completely separate objects.”

# In shallow copy: : inner list is SAME → so both changed
# In deep copy : inner list is NEW → so only copy changed
```

```python
l = [1, [2, [3, 4], 5], [6,8],3] # flatten

def flatten(l):
    res=[]
    for item in l:
        if isinstance(item,list):
            flatten(item)
        else:
            res.append(item)
flatten(l)
print(res)
```

```python
# 9️⃣ Find Common Elements
# Problem:Find common elements between two lists.
# Example:[1,2,3], [2,3,4] → [2,3]


def common_elemets():
    a=[1,2,3,5,6,5,4,6,7]
    b=[2,3,4,5,7]
    seen =set()
    res=[]
    for i in range(max(len(a),len(b))):
        if a[i] in b and i not in seen:
            res.append(a[i])
            seen.add(a[i])
    print(res)
            
    
common_elemets()
```

```python
# 8️⃣ Remove All Occurrences
# Problem:Remove all occurrences of a given element from a list.
# Example:[1,2,2,3], remove 2 → [1,3]

def remove_occurence():
    a=[1,2,2,3]
    occ=2
    res=[]
    for i in a:
        if i!=occ:
            res.append(i)
    
    print(res)
```

```python
# 5️⃣ Merge Two Lists Alternately
# Problem:Merge two lists by alternating elements.
# Example:[1,2] , [3,4] → [1,3,2,4]
 
def merge_array():
    a=[1,2,7,8]
    b=[3,4]
    result = []
    for i in range(max(len(a), len(b))):
        print(i)
        if i < len(a):
            result.append(a[i])
        if i < len(b):
            result.append(b[i])
    print(result)

merge_array()
```

```python

# 13️⃣ Find Minimum and Maximum
# Problem:Return both min and max from a list without using min() or max().
# Example:[3, 1, 5] → (1, 5)

def min_max():
    nums=[6,8,908,64,65]
    min=max=nums[0]
    
    for num in nums[1:]:
        if num<min:
            min=num
        elif num>max:
            max=num
    print(f"MIN : {min} === MAX : {max}")
min_max()
```
