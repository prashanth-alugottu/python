```python
# loc is the label based 
print(df)
print("=============================================")
print(df.loc[2,"Salary"])
```

```python
# loc is the label based 
print(df)
print("=============================================")
print(df.loc[3,"Salary"])
print(df.loc[0,"Salary"])
print(df.loc[2,"Salary"])
```

```python
print(df)
print("=============================================")
print(df.loc[1:3]) # in 'loc' Rows with labels 1, 2, 3 (end INCLUDED)
print("=============================================")
print(df.loc[1:3,"Salary"])
print("=============================================")
print(df.iloc[1:3]) # Rows 'iloc' index based from 1,2 
```

```python
# Filtering
print(df)
print("=============================================")
df['Age']>24
print("=============================================")
df[df['Age']>24]
```
