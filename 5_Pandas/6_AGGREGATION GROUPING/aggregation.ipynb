```python
import pandas as pd

data = {'Category': ['A','A','B','B','C'],
        'Sales': [100, 150, 200, 130, 90],
        'Profit': [20, 30, 50, 40, 15]}
df=pd.DataFrame(data)
df
```

```python
print(df)
print("=================================")
print(df.groupby('Category'))

print("\n====================== groupby Category and mean ======================")
mean=df.groupby('Category').mean()
print(mean)

print("\n====================== groupby Category and sum ======================")
df2=df.groupby('Category').sum()
print(df2)

print("\n====================== groupby Sales and sum ======================")
df3=df.groupby('Sales').sum() # since there are groups available
print(df3)

print("\n====================== groupby Category and count ======================")
df4=df.groupby('Category').count()
print(df4)

print("\n====================== groupby Sales and count ======================")
df5=df.groupby('Sales').count()
print(df5) # Now you can see the output why we cannt do groups since there are no matching values
```

```python
print(df)
print("\n================= highest Sales in each Category. =================")
df1=df.groupby('Category')['Sales'].max()
print(df1)

print("\n================= Lowest Sales in each Category. =================")
df2=df.groupby('Category')['Sales'].min()
print(df2)
```

```python
print(df)
print("\n================= average Sales per Category. =================")
df1=df.pivot_table(values='Sales', index='Category', aggfunc='mean')
print(df1)
```

```python
print(df)
print("\n================= Aggregate multiple columns. =================")
df1=df.agg({'Sales':'mean','Profit':'sum'})
print(df1)
```

```python
import numpy as np
print(df1)
df1=df.drop('Category',axis=1)
print(df1)
df1=df1.apply(np.mean)
print(df1)
```
