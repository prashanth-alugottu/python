```python
import pandas as pd
data = {
    'A': [10, 20, 30, 40, 50],
    'B': [5, 10, 15, 20, 25],
    'C': [2, 4, 6, 8, 10]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df)
print("\n============== 1.Mean (average value per column) ==============")
print(df.mean())
```

```python
print(df)
print("\n============== 2.Median (middle value per column) ==============")
print(df.median())
```

```python
print(df)
print("\n============== 3.Standard Deviation ==============")
print(df.std())
```

```python
print(df)
print("\n============== 4.Variance ==============")
print(df.var())
```

```python
print(df)
print("\n============== 5.Sum ==============")
print(df.sum())
```

```python
print(df)
print("\n============== 6.Minimum ==============")
print(df.min())
```

```python
print(df)
print("\n============== 7.Maximum ==============")
print(df.max())
```

```python
print(df)
print("\n============== 8.Count of Non-Null Values ==============")
print(df.count())
```

```python
print(df)
print("\n============== 9.Correlation Matrix ==============")
print(df.corr())
```
