```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva','banti'],
    'Age': [25, 30, 35, 40, 28,30],
    'Salary': [50000, 60000, 70000, 80000, 55000,90000],
    'Dept': ['HR', 'IT', 'Finance', 'IT', 'HR','gmarketing']
}

df=pd.DataFrame(data)
df
```

```python
print(df)
print("===================================================")
df1=df.sort_values('Age')
print(df1)
```

```python
print(df)
print("===================================================")
df2=df.sort_values('Dept')
print(df2)
```

```python
print(df)
print("\n====================== sort based on Salary ======================")
df3=df.sort_values('Salary',ascending=False)
# print(df3)
print("\n====================== sort based on Dept ======================")
df4=df.sort_values('Dept',ascending=False)
# print(df4)
print("\n====================== Multiple sorts ======================")
df5=df.sort_values(['Dept','Salary'],ascending=[True,False])
print(df5)
print("\n====================== Only sort with Dept descending ======================")
df6=df.sort_values('Dept',ascending=True)
print(df6)
```

```python
print("====================== Dataset ======================")
print(df)
print("\n====================== age>30 ======================")
df7=df[df['Age']>=30]
print(df7)
print("\n====================== busing query age>30 ======================")
df8=df.query('Age>=30')
print(df8)
print("\n====================== busing query Salary>50000 ======================")
df9=df.query('Salary>50000')
print(df9)
# print("\n====================== busing query Multi ======================")
# df10=df.query('Salary>50000 && Age>25')
# print(df10)
```

```python
print("====================== Dataset ======================")
print(df)
print("\n====================== Sample ======================")
df1=df.sample(2)
print(df1)
print("\n====================== nlarge -> Highest ======================")
df2=df.nlargest(3,'Salary')
print(df2)
# print("\n====================== nlarge -> in name ======================")
# df3=df.nlargest(2,'Name')
# print(df3)
print("\n====================== smalest ======================")
df4=df.nsmallest(2,'Salary')
print(df4)
```

```python
print("====================== Dataset ======================")
print(df)
print("\n====================== Filter columns a ======================")
df1=df.filter(like='a')
print(df1)
print("\n====================== Filter columns A ======================")
df2=df.filter(like='A')
print(df2)
print("\n====================== Filter columns Na ======================")
df3=df.filter(like='Na')
print(df3)
```
